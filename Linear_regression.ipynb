# Outliers Detection

```python
def outlier_detection(df):
    #data=[df[i] for i in df.columns]
    q1= np.percentile(df,25)
    q3=np.percentile(df,75)
    iqr = q3-q1
    low= q1-1.5*iqr
    high=q3+1.5*iqr
    #print(low)
    #print(high)
    vals= df[~((df<low)|(df>high)).any(axis=1)]
    #print(vals)
    #vals = df[~((df < low) |(df>high)).any(axis=1)]
    #boston_df_out.shape
    return vals
```

```python
#outlier_detection([df[i] for i in df.columns])
outlier_detection(df)

#new_df=pd.DataFrame()
#df.apply(outlier_detection,axis=1)
```

```python
boston_df_o1=df.copy()
Q1 = boston_df_o1.quantile(0.25)
Q3 = boston_df_o1.quantile(0.75)
IQR = Q3 - Q1
boston_df_out = boston_df_o1[~((boston_df_o1 < (Q1 - 1.5 * IQR)) |(boston_df_o1 > (Q3 + 1.5 * IQR))).any(axis=1)]
boston_df_out.shape
```
